## Neural Network function

```python
from tensorflow_addons import losses
```

```python
import tensorflow as tf
import keras_tuner as kt
from tensorflow import keras
from sklearn.metrics import confusion_matrix, classification_report
from keras.callbacks import EarlyStopping
from sklearn.metrics import auc,precision_recall_curve, PrecisionRecallDisplay
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

## Weight determination


```python
w_0 = 10015/(2*10000)
w_1 = 10015/(2*15)
```

```python
print(w_0)
print(w_1)
```

```python
weights = {0.0:w_0,1.0:w_1}
```

```python
def ANN(first_layer, dropout, second_layer,X_train, y_train,valid_set, X_test, y_test, loss,weights,name):
    model = keras.Sequential([
        keras.layers.Dense(first_layer, input_dim=11, activation='relu'),
        keras.layers.Dropout(dropout),
        keras.layers.Dense(second_layer, activation='relu'),
        keras.layers.Dropout(dropout),
        keras.layers.Dense(1, activation='sigmoid')
    ])
    METRICS = [
      
      keras.metrics.AUC(name='prc', curve='PR'), # precision-recall curve
    ]
    model.compile(optimizer='adam', loss=loss, metrics=METRICS)
    if weights == -1:
        model.fit(X_train, y_train, epochs=1000,validation_data = valid_set,  callbacks=EarlyStopping(monitor='val_loss',patience=10,verbose=1))
    else:
        model.fit(X_train, y_train, epochs=1000,validation_data = valid_set, class_weight = weights, callbacks=EarlyStopping(monitor='val_loss',patience=10,verbose=1))
     
       
    model.summary()
    
    print(model.evaluate(X_test, y_test))
    
    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)
    
    print("Classification Report: \n", classification_report(y_test, y_preds))
    
    print(y_test)
    print(y_preds)
    cm=confusion_matrix(y_test,y_preds)
    print(cm)
    cmhm = sns.heatmap(cm, annot=True)
    cmhm.set_xlabel("Predicted Value")
    cmhm.set_ylabel("True Value")
    cmhm.set_title(name)
    plt.savefig(name+"_cm", bbox_inches='tight')
    plt.show()
    
    # PR Curve
    # plot the model precision-recall curve
    precision, recall, thresholds = precision_recall_curve(y_test,y_preds)
    
    plt.plot(precision, recall, label=name, linewidth=2)
    plt.title(name)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.grid(True)
    ax = plt.gca()
    ax.set_aspect('equal')
    plt.legend(loc='lower right')
    plt.savefig(name+"_PRC", bbox_inches='tight')
    plt.show()
    
        
    auc_precision_recall = auc(recall, precision)
    print(auc_precision_recall)
    name=name
    
    return y_preds,auc_precision_recall,name
```

```python
AB_3_class0.shape
```

```python
def get_train_batch(df_majority, df_minority,start,end):
    df_train = pd.concat([df_majority[start:end],df_minority],axis=0)
    
    X_train = df_train.drop('MBC',axis='columns')
    y_train = df_train.MBC
    
    return X_train, y_train
```

```python
name = "Ensemble 1"
X_train, y_train = get_train_batch(AB_3_class0, AB_3_class1,0,11)

y_pred1,auc_ENS_1,sample_method_1 = ANN(8,0.2,4,X_train, y_train,valid_set, X_test, y_test, 'binary_crossentropy', -1,name)
```

```python
name = "Ensemble 2"
X_train, y_train = get_train_batch(AB_3_class0, AB_3_class1,12,23)

y_pred2, auc_ENS_2, sample_method_2 = ANN(8,0.2,4,X_train, y_train,valid_set, X_test, y_test, 'binary_crossentropy',-1,name)
```

```python
name = "Ensemble 3"
X_train, y_train = get_train_batch(AB_3_class0, AB_3_class1,24,35)

y_pred3, auc_ENS_2, sample_method_2 = ANN(8,0.2,4,X_train, y_train,valid_set, X_test, y_test, 'binary_crossentropy',-1,name)
```

```python
y_pred_final = y_pred1.copy()

for i in range(len(y_pred1)):
    n_ones = y_pred1[i] + y_pred2[i] + y_pred3[i]
    if n_ones >1:
        y_pred_final[i] = 1
    else:
        y_pred_final[i] = 0
    
```

```python
print(classification_report(y_test,y_pred_final))

name = "Ensemble_Undersampling"
    
print(y_test)
print(y_pred_final)
cm=confusion_matrix(y_test,y_pred_final)
print(cm)
cmhm = sns.heatmap(cm, annot=True)
cmhm.set_xlabel("Predicted Value")
cmhm.set_ylabel("True Value")
cmhm.set_title(name)
plt.savefig(name+"_cm", bbox_inches='tight')
plt.show()
# PR Curve
# plot the model precision-recall curve
precision, recall, _ = precision_recall_curve(y_test,y_pred_final)

plt.plot(precision, recall, label=name, linewidth=2)
plt.title(name)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.grid(True)
ax = plt.gca()
ax.set_aspect('equal')
plt.legend(loc='lower right')
plt.savefig(name+"_PRC", bbox_inches='tight')
plt.show()
    
    
    

    
auc_precision_recall_ens = auc(recall, precision)
print(auc_precision_recall_ens)
results = results.append({'Method':name,'PR AUC':auc_precision_recall_ens}, ignore_index= True)
```
